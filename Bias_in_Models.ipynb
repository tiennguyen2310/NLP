A bias may occur concerning differences in information that models can scrape through.

```python
from transformers import pipeline
unmasker = pipeline("fill-mask", model="bert-base-uncased")
```

# **Task :** Fill the word to the sentence.

Notice the difference when the subject is `man` and `woman`.

```python
result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])
```

```python
result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```

# **Result Analysis**

`The only gender-free answer is waiter/waitress!`

`=> The original model could very easily generate sexist, racist, or homophobic content.`
